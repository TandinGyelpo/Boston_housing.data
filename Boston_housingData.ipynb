```python
X=df['RM'].values.reshape(-1,1)
```

```python
y=df['MEDV'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
```

```python
LR.fit(X_train,y_train)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
yp=LR.predict(X_test)
```

MSC

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,yp)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from statsmodels.tools.eval_measures import rmse
MAE= mean_absolute_error(y_test,yp)
MSE= mean_squared_error(y_test,yp)
r2 =r2_score(y_test,yp)
RMSE= rmse(y_test,yp)

print("MAE=",MAE,"\nMSE=",MSE,"\nr2=",r2,"\nRMSE=",RMSE)
```

# Assumption of Regression Analysis
1. Linear Relationship between the feature and target :Linear relationship between the feature and target
2. Little or no Multicollinearity between the features: Multicollinearity is a state very high inter-correlation or inter-associates among the independent variables. it is therefore a type of distribance in data present weaknes the statictical power of the regression model. Pair plots and heatmaps(correlation matrix) can be used for identifying highly correlated features.
3. No or low Homoscedasticity : error term should not explain other error term behaviours when ploted scatterplot,should show randomly spread instance of any pattern.
4. Normal Distribution of error
5. Little or No autocorrelation in the residuals:
